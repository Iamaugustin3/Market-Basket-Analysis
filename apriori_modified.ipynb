```python
def scanning(dataset, candidates, min_support):
    st = {}
    print("reached scanning")
    
    for i in dataset:
        print("value in scanning dataset is=============================",i)
        
    print("The candidate data is ===",candidates)
    num_items = float(len(list(dataset)))
    print("no.of items",num_items)
    for t in dataset:
        print("printing t",t)
        for c in candidates:
            print("printing c",c)
            if c.issubset(t):
                st.setdefault(c, 0)
                
                st[c] += 1

    rlist = []
    support_data = {}
    for key in st:
        support = st[key] / num_items
        print("support:",support)
        if support >= min_support:
            rlist.insert(0, key)
            support_data[key] = support
    print ("list after scanning: {}".format(rlist))
    print ("support dictionary: {}".format(support_data))
    print ("---------------")
    return rlist, support_data
```

```python
def Generation(frequent, m):
    rList = []
    lengthAm = len(frequent)
    #print ("length of Am",lengthAm)
    for i in range(lengthAm):
        for j in range(i + 1, lengthAm):
            print("frequent[i]",frequent[i])
            A1 = list(frequent[i])[:m - 2]
            print ("L1=",A1)
            A2 = list(frequent[j])[:m - 2]
            print ("L2=",A2)
            A1.sort()
            A2.sort()
            if A1 == A2:
                rList.append(frequent[i] | frequent[j])
                #print("rList...",rList)
    print ("list after generation:",rList)
    return rList
```

```python
def Apriori(minimumsupport,dataset):
    print("The dataset on apriori is ",dataset)
    c1 = []
    #creating subset of individual item 
    for t in dataset:
        for item in t:
            if not [item] in c1:
                c1.append([item])

    print("The c1 before sort is ",c1)
    c1.sort()
   
    C1=list(map(frozenset, c1))
    print("This is the frozen set of C1 = \n")
    print(C1)

    #using set funtion to remove duplicates
    D = list(map(set, dataset))
    print("printing D= \n",D)
    print("step to call scanning function")
    A1, support_data = scanning(D, C1, minimumsupport)
    print("::::::::",support_data)
    A = [A1]
    print("A -------------------",A)
    k = 2
    while (len(A[k - 2]) > 0):
        
        Ck = Generation(A[k - 2], k)      
        Ak, supK = scanning(D, Ck, minimumsupport)
        support_data.update(supK)
        A.append(Ak)        
        k += 1
        print("---------------FREQUENT SETS--------------------")
        print ("frequent sets::: ",Ak)
        print ("----------------------------------")
    
    return A, support_data
```
